```python
import matplotlib.pyplot as plt
import networkx as nx 
```

```python
G = nx.barabasi_albert_graph ( n=10, m=4, seed=None, initial_graph=None )
print ( G )
```

```python
G = nx.watts_strogatz_graph ( n=10, k=6, p=0.2, seed=None )
print ( G )
```

```python
G = nx.random_tree( n=20 )
print ( G )
```

```python
layout = nx.spring_layout ( G )
# layout = nx.planar_layout ( G )           # only if G is planar else Network Exception
# layout = nx.circular_layout ( G )

fig, ax = plt.subplots(figsize=(8, 8))
nx.draw(
  G,
  pos=layout,
  with_labels=False,
  ax=ax,
)
```
